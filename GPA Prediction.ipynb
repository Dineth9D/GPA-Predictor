```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression,BayesianRidge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn import ensemble
from sklearn.neighbors import KNeighborsRegressor
from sklearn import neighbors
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
```

```python
#Nominal Encoding
def Nominal_encoder(model):
    features=model.columns
    for i in range(len(features)):
        model[features[i]]=model[features[i]].replace(['A+','A','A-','B','B+','B-','C','C+','C-','D','D+','F','W','WU','I','Ab'], [4,4,3.7,3,3.4,2.7,2,2.4,1.7,1,1.4,0,0,0,0,0])
```

***Here we get only 12 rows only***

```python
grade_dataset = grade_dataset.drop(["Seat No.", "HS-205/20","MT-222","EE-222","MT-224","CS-210","CS-211","CS-203","CS-214","CS-212","EE-217","CS-215","CS-301","CS-302","MT-331","EF-303","HS-304","TC-383","MT-442","EL-332","CS-318","CS-306","CS-312","CS-317","CS-403","CS-421","CS-406","CS-414","CS-419","CS-423","CS-412"], axis=1)
```

```python
X_m1 =grade_dataset.values[:, 0:11]
Y_m1 =grade_dataset.values[:, 11]
```


**Algorithm 1: Linear Regression**

```python
#Train Test Splitting 
x_train,x_test,y_train,y_test=train_test_split(X_m1,Y_m1,test_size=0.2,random_state=3)

def model1Algo1(mode):
    if (mode=='train'):
        LR = LinearRegression()
        LR.fit(x_train,y_train)
        y_pred=LR.predict(x_test)

        print("Mean Squared Error: ",mean_squared_error(y_test,y_pred))
        print("Training data R-squared:", LR.score(x_train,y_train))
        print("Test data R-squared:", LR.score(x_test,y_test))

        print(LR.predict([[1,1,4,3.7,3.4,1,1,1,1,1,3]]))
        print(LR.predict([[4,4,4,4,4,4,4,4,4,4,4]]))
    elif (mode=='test'):
        LR=LinearRegression()
        LR.fit(X_m1,Y_m1)
        return LR
    else:
        print("Invalid args")
model1Algo1("train")
```


**Algorithm 2: Polynomial Regression**

```python
#Train Test Splitting 
x_train,x_test,y_train,y_test=train_test_split(X_m1,Y_m1,test_size=0.2,random_state=3)

def model1Algo2(mode):
    if (mode=='train'):
        deg=2   #if try 3 and 4 model will over fit
        X_train_poly2=PolynomialFeatures(degree=deg).fit_transform(x_train)
        X_train_poly2=pd.DataFrame(X_train_poly2)
        X_train_poly2=X_train_poly2.drop([0],axis=1)

        PM=LinearRegression()
        PM.fit(X_train_poly2,y_train)

        print("Training data R-squared:", PM.score(X_train_poly2, y_train))

        X_test_poly2=PolynomialFeatures(degree=deg).fit_transform(x_test)
        X_test_poly2=pd.DataFrame(X_test_poly2)
        X_test_poly2=X_test_poly2.drop([0],axis=1)
        y_pred=PM.predict(X_test_poly2)

        print("Test data R-squared:", PM.score(X_test_poly2,y_test))
        print("Number of non-zero features:",np.sum(PM.coef_ != 0))
        print("Mean Squared Error:",mean_squared_error(y_test,y_pred))

        test=PolynomialFeatures(degree=2).fit_transform([[4,2.7,1,3,1,2,1,3.7,4,4,1]])
        test=pd.DataFrame(test)
        test=test.drop([0],axis=1)
        print(PM.predict(test))
    elif (mode=='test'):
        deg=2   #if try 3 and 4 model will over fit
        modified_x=PolynomialFeatures(degree=deg).fit_transform(X_m1)
        modified_x=pd.DataFrame(modified_x)
        modified_x=modified_x.drop([0],axis=1)
        PM=LinearRegression()
        PM.fit(modified_x,Y_m1)
        return PM
    else:
        print("Invalid Args")
model1Algo2('train')
```


**Algorithm 3: Bayesian Ridge Regression**

```python
#Train Test Splitting 
x_train,x_test,y_train,y_test=train_test_split(X_m1,Y_m1,test_size=0.2,random_state=3)
# the above test_size is selected by hit and trial by first splitting test_size in range from 0.1- 0.9
# we can have any random_state but I need to fix it for a particular model's all algorithms so that I can find the best algorithm
def model1Algo3(mode):
    if (mode=='train'):
        BayReg =BayesianRidge(fit_intercept=True)
        BayReg.fit(x_train,y_train)
        Bay_pred = BayReg.predict(x_test)

        score=r2_score(y_test,Bay_pred )

        print("R2 Score",score)
        print("Training data R-squared for Baysian:", BayReg.score(x_train,y_train))
        print("Mean Squared Error:",mean_squared_error(y_test,Bay_pred ))
        print(BayReg.predict([[2,3.4,0,1.4,1.7,2.7,0,1,3.7,3,4]])) # randomly predicted model on these inputs
    elif (mode=='test'):
        BayReg =BayesianRidge(fit_intercept=True)
        BayReg.fit(X_m1,Y_m1)
        return BayReg
    else:
        print("Invalid Args")
model1Algo3('train')
```
